## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

###Playing with Discrete Distributions

**Task9**:

Plot the pdfs (techincally the pmfs) of the following distributions: $Bernoulli(p=0.7)$, $Binomial(n=10,p=0.7)$, $Poisson(λ=7)$, $Geometric(p=0.7)$, $HyperGeometric(N=50,k=35,n=15)$, $Categorical(p=[0.05,0.3,0.05,0.1,0.2,0.4])$

Additionally, compute the first four moments for all those distributions--except the Categorical distribution.

````python
pz.Bernoulli(p=0.7).plot_pdf()
print(pz.Bernoulli(p=0.7).moments(types='mvsk'))
pz.Binomial(n=10,p=0.7).plot_pdf()
print(pz.Binomial(n=10,p=0.7).moments(types='mvsk'))
pz.Poisson(mu=7).plot_pdf()
print(pz.Poisson(mu=7).moments(types='mvsk'))
pz.Geometric(p=0.7).plot_pdf()
print(pz.Geometric(p=0.7).moments(types='mvsk'))
pz.HyperGeometric(N=50,k=35,n=15).plot_pdf()
print(pz.HyperGeometric(N=50,k=35,n=15).moments(types='mvsk'))
pz.Categorical(p=[0.05,0.3,0.05,0.1,0.2,0.4]).plot_pdf()
````

**Task10**:

Describe what processes "generate" each of the previous discrete distributions models. Also report their support.

Here's a spicier way to phrase the question: describe these distributions' maximum entropy conditions in english.

Hint: Make sure to read any of the: preliZ documentation, relevant wikipedia articles, or wolfram alpha.


**Answer**

Bernoulli (p=0.7): This distribution model has a single trial with two possible outcomes: success (p=.7) and failure (1-p=.3). Support: {0,1}

Binomial(n=10,p=0.7): This distribution models the number of successes in 10 independent trials, and all of them have a success probability of p=.7. Support: {0,1,3,4,5,6,7,8,9,10}

Poisson(λ=7): This distribution models the number of events that occur in a fixed interval of time, where these events happen independently and at a constant average rate λ=7. Supoort: {0,∞}

Geometric (p=0.7): This distribution models the number of trials needed to get the first success in a series of independent trials with success probablility p=0.7. Support: the number of trials from the first success, starting from 1

Hypergeometric(N=50,k=35,n= 15): This distribution models the number of successes in a sample of size of 15 drawn without replacement from a population size of 50 with k=35 successes.

Categorical(p=[0.05,0.3,0.05,0.1,0.2,0.4]) This distribition models a single trial that results in one of categories given, with specified for each category. Support:The set of categories
